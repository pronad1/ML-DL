### Demo Classification with multilayer perceptron

```python
# Import nessesary library

import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_circles
```

```python
# Generate synthetic data (circles)

X,y=make_circles(n_samples=300,noise=0.1,factor=0.5,random_state=0)
```

```python
# Plot the points with labels
plt.Figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.RdYlBu,edgecolors='k',marker='o',s=50)
plt.xlabel('Feature 1 (X[:,0])')
plt.ylabel('Feature 2 (X[:,1])')
plt.title('Plot ot Points with Labels (make_circles Dataset)')
plt.colorbar(label='Class Label')
plt.show()
```
