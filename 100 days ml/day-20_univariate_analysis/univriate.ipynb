### Univriate
- When we analysis examines a single variable at a time to describe its distribution, central tendencies (mean, median, mode), and spread (range, standard deviation). 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

### 1. Catagorical Data
a. Countplot

```python
#sns.countplot(x='Pclass', data=df)
sns.countplot(x='Survived', data=df)
#df['Survived'].value_counts().plot(kind='bar')
```

b. Piechart

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f%%')
```

### 2. Numerical Data
a.Histogram

```python
plt.hist(df['Age'].dropna(), bins=30)                           
```

b. Distplot

```python
sns.displot(df['Age'].dropna(), kde=True)
```

c. Boxplot

![Boxplot Example](image.png)

```python
# Five number summary that help to find out potential outliers
# Minimum, Q1, Median, Q3, Maximum
```
