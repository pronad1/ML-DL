```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().plot.bar()
```

```python
df['GarageQual'].mode()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde', ax=ax)

df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde', ax=ax, color='red')

line, labels = ax.get_legend_handles_labels()
labels = ['House with TA', 'House with NA']
ax.legend(line, labels, loc='best')

plt.title('GarageQual')
```

```python
temp = df[df['GarageQual'].isnull()]['SalePrice']
```

```python
df.fillna({'GarageQual': 'TA'}, inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

temp.plot(kind='kde', ax=ax, color='red')

# distribution of the variable after imputaion
df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde', ax=ax)

line, labels = ax.get_legend_handles_labels()
labels = ['Original Variable', 'Imputed Variable']

ax.legend(line, labels, loc='best')
plt.title('GarageQual')
```
